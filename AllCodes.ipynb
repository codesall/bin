```python
import tkinter as tk
from tkinter import *

# Create the main window
root = tk.Tk()
root.title("Student Registration Form")
root.geometry("400x500")

# Title Label
title_label = tk.Label(root, text="Student Registration Form", font=("Arial", 16, "bold"))
title_label.pack(pady=10)

# Name Label and Entry
name_label = tk.Label(root, text="Full Name:", font=("Arial", 10))
name_label.pack(anchor="w", padx=20, pady=5)
name_entry = tk.Entry(root, width=40)
name_entry.pack(padx=20, pady=5)

# Age Label and Entry
age_label = tk.Label(root, text="Age:", font=("Arial", 10))
age_label.pack(anchor="w", padx=20, pady=5)
age_entry = tk.Entry(root, width=40)
age_entry.pack(padx=20, pady=5)

# Gender Label and Radio Buttons
gender_label = tk.Label(root, text="Gender:", font=("Arial", 10))
gender_label.pack(anchor="w", padx=20, pady=5)

gender_var = tk.StringVar()

male_radio = tk.Radiobutton(root, text="Male", variable=gender_var, value="Male")
male_radio.pack(anchor="w", padx=40)

female_radio = tk.Radiobutton(root, text="Female", variable=gender_var, value="Female")
female_radio.pack(anchor="w", padx=40)

other_radio = tk.Radiobutton(root, text="Other", variable=gender_var, value="Other")
other_radio.pack(anchor="w", padx=40)

# Courses Label and Checkbuttons
courses_label = tk.Label(root, text="Courses Interested In:", font=("Arial", 10))
courses_label.pack(anchor="w", padx=20, pady=5)

python_var = tk.BooleanVar()
python_cb = tk.Checkbutton(root, text="Python", variable=python_var)
python_cb.pack(anchor="w", padx=40)

java_var = tk.BooleanVar()
java_cb = tk.Checkbutton(root, text="Java", variable=java_var)
java_cb.pack(anchor="w", padx=40)

ml_var = tk.BooleanVar()
ml_cb = tk.Checkbutton(root, text="Machine Learning", variable=ml_var)
ml_cb.pack(anchor="w", padx=40)

# Country Label and Dropdown List
country_label = tk.Label(root, text="Country:", font=("Arial", 10))
country_label.pack(anchor="w", padx=20, pady=5)

country_list = ["United States", "Canada", "India", "Australia", "United Kingdom"]
country_var = tk.StringVar(value=country_list[0])

country_dropdown = ttk.Combobox(root, textvariable=country_var, values=country_list, width=37)
country_dropdown.pack(padx=20, pady=5)

# Address Label and Textbox
address_label = tk.Label(root, text="Address:", font=("Arial", 10))
address_label.pack(anchor="w", padx=20, pady=5)

address_text = tk.Text(root, width=30, height=4)
address_text.pack(padx=20, pady=5)

# Submit Button
def submit_form():
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_var.get()
    country = country_var.get()
    address = address_text.get("1.0", "end-1c")
    courses = []
    if python_var.get(): courses.append("Python")
    if java_var.get(): courses.append("Java")
    if ml_var.get(): courses.append("Machine Learning")

    # Here you can handle form submission, e.g., saving data to a file or database
    print("Submitted Data:")
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Gender: {gender}")
    print(f"Country: {country}")
    print(f"Address: {address}")
    print(f"Courses: {', '.join(courses)}")

submit_button = tk.Button(root, text="Submit", command=submit_form)
submit_button.pack(pady=20)

# Run the application
root.mainloop()
```
